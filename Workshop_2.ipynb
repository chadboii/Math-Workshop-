## Exercise 1: Distinct elements from a list

```python
def unique_list(items):
    """Return a new list containing only distinct elements while keeping order."""
    unique = []
    for item in items:
        if item not in unique:
            unique.append(item)
    return unique

sample_list = [1, 2, 3, 3, 3, 3, 4, 5]
print("Sample List:", sample_list)
print("Unique List:", unique_list(sample_list))
```

## Exercise 2: Check whether a number is prime

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

## Exercise 3: Twin primes less than 1000

```python
def print_twin_primes(limit):
    """Print twin prime pairs less than the given limit."""
    for n in range(3, limit - 1, 2):
        if is_prime(n) and is_prime(n + 2):
            print((n, n + 2))

print("Twin primes less than 1000:")
print_twin_primes(1000)
```

## Exercise 4: Sum of divisors excluding the number itself

```python
def sumOfDivisors(n):
    """Return sum of all divisors of n excluding n itself."""
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

print("Sum of divisors of 6 excluding 6:", sumOfDivisors(6))
```

## Exercise 5: Find the next perfect number after 6

```python
def next_perfect_number_after_6():
    """Find and return the next perfect number after 6."""
    n = 7
    while True:
        if sumOfDivisors(n) == n:
            return n
        n += 1

print("The next perfect number after 6 is:", next_perfect_number_after_6())
```

## Exercise 6: Mean and median without library function

```python
Marks = [20, 30, 35, 55, 70, 80, 90]

# Mean = total sum divided by number of values
total = 0
count = 0
for mark in Marks:
    total += mark
    count += 1
mean = total / count

# Median = middle value after sorting
sorted_marks = sorted(Marks)
n = len(sorted_marks)
middle = n // 2

if n % 2 == 1:
    median = sorted_marks[middle]
else:
    median = (sorted_marks[middle - 1] + sorted_marks[middle]) / 2

print("Mean:", mean)
print("Median:", median)
```

## Exercise 7: Convert decimal number to binary number

```python
def decimal_to_binary(decimal):
    """Convert a decimal number to binary without using bin()."""
    if decimal == 0:
        return "0"

    binary = ""
    n = abs(decimal)

    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2

    if decimal < 0:
        binary = "-" + binary

    return binary

num = int(input("Enter a decimal number: "))
print("Binary number:", decimal_to_binary(num))
```

## Exercise 8: Product of digits

```python
def prodDigits(number):
    """Return the product of all digits of a number."""
    number = abs(int(number))

    if number == 0:
        return 0

    product = 1
    while number > 0:
        digit = number % 10
        product *= digit
        number = number // 10
    return product

num = int(input("Enter a number: "))
print("Product of digits:", prodDigits(num))
```

## Output of the `capitals(**kwargs)` function

```python
def capitals(**kwargs):
    print(kwargs)
    print()
    for (key, value) in kwargs.items():
        print(key, "->", value)

capitals(Nepal="Ktm", India="Delhi", China="Beijing", France="Paris")
```

## Exercise 9: Multiply unknown number of arguments using `*args`

```python
def multiply_all(*args):
    """Multiply all numbers passed as arguments."""
    product = 1
    for number in args:
        product *= number
    return product

print("Product of 1, 2, 3, 4, 5:", multiply_all(1, 2, 3, 4, 5))
print("Product of 12, 13, 14:", multiply_all(12, 13, 14))
```
